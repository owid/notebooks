```python
import pandas as pd
```

```python
raw = pd.read_csv("inputs/fishstat.csv")
countries = pd.read_csv("inputs/countries.csv")
gear = pd.read_csv("inputs/gear_type.csv")
```

```python
raw
```

```python
#By end use
```

```python
global_end_use = raw.groupby(["end_use_type", "year"])["tonnes"].agg('sum').reset_index()
global_end_use["fishing_entity"]="World"
end_use = raw.groupby(["year", "fishing_entity","end_use_type", ])["tonnes"].agg('sum').reset_index()
```

```python
global_end_use = global_end_use.pivot_table("tonnes", ["year", "fishing_entity"],"end_use_type").reset_index()
end_use = end_use.pivot_table("tonnes", ["year", "fishing_entity"],"end_use_type").reset_index()
```

```python
end_use = pd.concat([end_use, global_end_use])
```

```python
end_use["total_catch"] = end_use["Direct human consumption"] + end_use["Fishmeal and fish oil"] + end_use["Other"]
```

```python
#By gear type
```

```python
global_gear_type = raw.groupby(["gear_type", "year"])["tonnes"].agg('sum').reset_index()
global_gear_type["fishing_entity"]="World"
gear_type = raw.groupby(["year", "fishing_entity","gear_type"])["tonnes"].agg('sum').reset_index()
```

```python
global_gear_type = pd.merge(global_gear_type, gear)
gear_type = pd.merge(gear_type, gear)
```

```python
gear_type 
```

```python
global_gear_type = global_gear_type.groupby(["year", "fishing_entity","gear"])["tonnes"].agg('sum').reset_index()
gear_type = gear_type.groupby(["year", "fishing_entity","gear"])["tonnes"].agg('sum').reset_index()
```

```python
global_gear_type = global_gear_type.rename(columns={"gear":"gear_type"})
gear_type = gear_type.rename(columns={"gear":"gear_type"})                                         
```

```python
gear_type
```

```python
global_gear_type = global_gear_type.pivot_table("tonnes", ["year", "fishing_entity"],"gear_type").reset_index()
gear_type = gear_type.pivot_table("tonnes", ["year", "fishing_entity"],"gear_type").reset_index()
```

```python
gear_type = pd.concat([gear_type, global_gear_type])
```

```python
gear_type
```

```python
#By sector
```

```python
global_fishing_sector = raw.groupby(["fishing_sector", "year"])["tonnes"].agg('sum').reset_index()
global_fishing_sector["fishing_entity"]="World"
fishing_sector = raw.groupby(["year", "fishing_entity","fishing_sector"])["tonnes"].agg('sum').reset_index()
```

```python
global_fishing_sector = global_fishing_sector.pivot_table("tonnes", ["year", "fishing_entity"],"fishing_sector").reset_index()
fishing_sector = fishing_sector.pivot_table("tonnes", ["year", "fishing_entity"],"fishing_sector").reset_index()
```

```python
fishing_sector = pd.concat([fishing_sector, global_fishing_sector])
```

```python
#By catch type
```

```python
global_catch_type = raw.groupby(["catch_type", "year"])["tonnes"].agg('sum').reset_index()
global_catch_type["fishing_entity"]="World"
catch_type = raw.groupby(["year", "fishing_entity","catch_type"])["tonnes"].agg('sum').reset_index()
```

```python
global_catch_type = global_catch_type.pivot_table("tonnes", ["year", "fishing_entity"],"catch_type").reset_index()
catch_type = catch_type.pivot_table("tonnes", ["year", "fishing_entity"],"catch_type").reset_index()
```

```python
catch_type = pd.concat([catch_type, global_catch_type])
```

```python
landings = raw[raw["catch_type"] == "Landings"]
```

```python
global_reported_landings = landings.groupby(["reporting_status", "year"])["tonnes"].agg('sum').reset_index()
global_reported_landings["fishing_entity"]="World"
reported_landings = landings.groupby(["year", "fishing_entity","reporting_status"])["tonnes"].agg('sum').reset_index()
```

```python
global_reported_landings = global_reported_landings.pivot_table("tonnes", ["year", "fishing_entity"],"reporting_status").reset_index()
reported_landings = reported_landings.pivot_table("tonnes", ["year", "fishing_entity"],"reporting_status").reset_index()
```

```python
reported_landings = pd.concat([reported_landings, global_reported_landings])
```

```python
combined = pd.merge(end_use, gear_type, how="outer")
```

```python
combined = pd.merge(combined, fishing_sector, how="outer")
```

```python
combined = pd.merge(combined, catch_type, how="outer")
```

```python
combined = pd.merge(combined, reported_landings, how="outer")
```

```python
combined = pd.merge(combined, countries)
```

```python
combined = combined.rename(columns = {"year":"Year",
                                      "Direct human consumption":"fish_direct_human_consumption",
                                      "Fishmeal and fish oil":"fishmeal_and_oil",
                                       "Other":"fish_other_uses",
                                       "artisinal fishing":"artisinal_fishing",
                                       "Landings":"landings",
                                       "Discards":"discards",
                                       "Artisinal":"artisinal",
                                       "Industrial":"industrial",
                                       "Recreational":"recreational",
                                       "Subsistence":"subsistence",
                                       "Reported":"reported_landings",
                                       "Unreported":"unreported_landings"
                                      }
                           )
```

```python
combined = combined.drop(columns=["fishing_entity"])
```

```python
combined = combined[ [ col for col in combined.columns if col != "Entity" ] + ["Entity"] ]
combined = combined[ ["Entity"] + [ col for col in combined.columns if col != "Entity" ] ]
```

```python
combined.to_csv("output/Global fish catch (SeaAroundUs).csv", index=False)
```
